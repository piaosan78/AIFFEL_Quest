```python
import random as r

class Account:                         # 클래스 생성
    account_count = 0               # 생성된 은행 계좌의 개수를 카운트 하는 곳

    def __init__(self, name, balance):      # 매직 메서드로 사용자와 초기 잔액 입력
        self.bank = "SC은행"
        self.name = name
        self.balance = balance
        self.account_number = str(r.randint(0,999)).zfill(3) + '-' + str(r.randint(0,99)).zfill(2) + '-' + str(r.randint(0,999999)).zfill(6)   # randint로 범위 숫자 랜덤 생성, zfill로 빈 자리에 0 채우기
        self.deposit_count = 0
        self.deposit_record = []
        self.withdraw_record = []
        Account.account_count += 1


# 클래스 변수 출력 : Account 클래스로부터 생성된 계좌의 개수를 출력하는 get_account_num() 메서드를 추가하세요.
    def get_account_num(self):
        return f'생성된 계좌 개수는 {self.account_count}개 입니다.'

# 입금 메서드 : Account 클래스에 입금을 위한 deposit 메서드를 추가하세요. 입금은 최소 1원 이상만 가능합니다.
    def deposit(self):
        print()
        d = int(input('입금하실 금액을 입력해주세요. : '))
        if d >= 1 :
            self.deposit_count += 1
            self.balance += d
            self.deposit_record.append(d)
            print(f'입금하실 금액은 {d:,}원입니다')

        while True:
            print()
            i = int(input('추가 입금을 원하시면 1번을 종료하시려면 0번을 입력해주세요. : '))
            if i == 1 :
                d1 = int(input('입금하실 금액을 입력해주세요. : '))
                self.deposit_count += 1
                self.balance += d1
                self.deposit_record.append(d1)
                print()
                print(f'입금하실 금액은 {d1:,}원입니다')
            elif i == 0 :
                break



# 출금 메서드 : Account 클래스에 출금을 위한 withdraw 메서드를 추가하세요. 출금은 계좌의 잔고 이상으로 출금할 수는 없습니다.
    def withdraw(self):
        print()
        w = int(input('출금하실 금액을 입력해주세요. : '))
        if (self.balance - w) < 0  :
            print('잔고 이상으로 출금할 수는 없습니다')
        else :
            self.balance -= w
            self.withdraw_record.append(w)
            print(f'{w:,}원이 출금되었습니다')

        while True :
            print()
            o = int(input('추가 출금을 원하시면 1번을 종료하시려면 0번을 입력해주세요. : '))
            if o == 1 and (self.balance - w) < 0 :
                w1 = int(input('출금하실 금액을 입력해주세요. : '))
                self.balance -= w1
                self.withdraw_record.append(w1)
                print(f'출금하실 금액은 {w1:,}원입니다')
            elif o == 0 :
                break


# 정보 출력 메서드 : Account 인스턴스에 저장된 정보를 출력하는 display_info() 메서드를 추가하세요. 잔고는 세자리마다 쉼표를 출력하세요.(은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고: 10,000원)
    def display_info(self) :
        print()
        print(f'은행이름: {self.bank}, 예금주: {self.name}, 계좌번호: {self.account_number}, 잔고: {self.balance:,}원')


# 이자 지급하기 입금 횟수가 5회, 10회, 15회가 될 때 잔고를 기준으로 1%의 이자가 잔고에 추가되도록 코드를 변경해보세요.
    def interest(self):
        if self.deposit_count == 5 or 10 or 15 :
            self.balance += (self.balance * 0.01)

#MainQuest Q10. 입출금 내역
#입금과 출금 내역이 기록되도록 코드를 업데이트 하세요.
#입금 내역과 출금 내역을 출력하는 deposit_history와 withdraw_history 메서드를 추가하세요.
    def deposit_history(self) :
        for x in self.deposit_record :
            print()
            print(f'{x}원 입금')

    def withdraw_history(self) :
        for y in self.withdraw_record :
            print()
            print(f'{y}원 출금')

# 여러 객체 생성 Account 클래스로부터 3개 이상 인스턴스를 생성하고 생성된 인스턴스를 리스트에 저장해보세요.
detail = []

account1 = Account('김이나',20000000)
account2 = Account('박이해',5000000)
account3 = Account('장군',980000)

detail.append(account1)
detail.append(account2)
detail.append(account3)

# 객체 순회 반복문을 통해 리스트에 있는 객체를 순회하면서 잔고가 100만원 이상인 고객의 정보만 출력하세요.
for i in detail :
    if i.balance >= 1000000:
        i.display_info()


bank = Account('박이해',10000)
bank.deposit()
bank.withdraw()
bank.interest()
bank.display_info()
bank.deposit_history()
bank.withdraw_history()
```
