```python
import re
name = "CharizardMega Charizard X"
name_split = name.split(" ")
name_split
# name을 공백 기준으로 분할하여 변수에 기록한다.
```

```python
temp = name_split[0]
temp
# 분할된 문자열에서 인덱스 0번의 문자만 분할해서 변수에 기록한다.
```

```python
tokens = re.findall('[A-Z][a-z]*', temp)
tokens
# 대문자 A-Z로 시작하고 그 뒤에 소문자 a-z가 하나 이상 반복적으로 나오는 모든 문자 패턴을
# temp 안에서 찾고 변수에 기록한다.
```

```python
tokens = []
for part_name in name_split:
    a = re.findall('[A-Z][a-z]*', part_name)
    tokens.extend(a)
tokens
# tokens 라는 빈 리스트를 만들고 반복목을 통해 name_split에 들어있는 분할된 문자들을 차례로 불러와
# part_name 할당하며 불러온 횟수만큼 반복한다.
# 할당된 part_name 을 re.findall 정규 표현식을 적용하여 a 변수에 기록한다.
# tokens 라는 변수에 extend 라는 메서드를 사용하여 a에 할당된 내용을 원본에 변경하며 적용한다.
```

```python
# 함수 데이터셋 작성 ※ name.split을 지정안할 경우 위에 있는 "CharizardMega Charizard X"을 불러오는 문제 발생함
def tokenize(name):
    name_split = name.split(" ")
    tokens = []
    for part_name in name_split:
        a = re.findall('[A-Z][a-z]*', part_name)
        tokens.extend(a)
    return np.array(tokens)
```

```python
all_tokens = list(legendary["Name"].apply(tokenize).values)

token_set = []
for token in all_tokens:
    token_set.extend(token)

print(len(set(token_set)))
print(token_set)
# 바로 위쪽 코드와 연계해서 코드 리뷰 필수
```
