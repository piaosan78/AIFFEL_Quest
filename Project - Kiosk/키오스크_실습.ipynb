```python
class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]

# 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ':', self.price[i], '원')

# 주문 메서드
    def menu_select(self):
        self.order_menu = []       # 추가 주문리스트
        self.order_price = []      # 추가 주문 금액
        print()    # 줄 바꿈
        n = int(input("음료 번호를 입력하세요 : "))   # 음료 번호 입력
        self.price_sum = self.price[n-1]   # 선택 음료의 가격

        # 음료 온도 물어보기
        t = 0   # 기본값 입력
        while t != 1 and t != 2 :
            t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
            if t == 1 :
                self.temp = "HOT"
            elif t == 2 :
                self.temp = "ICE"
            else :
                print("다시 입력하세요.\n")
        print('추가 주문 음료', self.temp, self.menu[n-1], ' : ', self.price[n-1], '원')

    # 추가 주문 메서드
        while n != 0 :    # 지불을 선택하기 전까지 반복
            print()    # 줄 바꾸기
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))      # 추가 주문 또는 지불
            if n > 0 and n < len(self.menu) + 1 :
                # 추가 음료 온도
                t = 0
                while t != 1 and t != 2 :
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1 :
                        self.temp = "HOT"
                    elif t == 2 :
                        self.temp = "ICE"
                    else :
                        print("다시 입력하세요.\n")
                self.order_price.append(self.price[n-1])      # 가격 리스트에 추가
                self.order_menu.append(self.temp + ' ' + self.menu[n-1])      # 주문 리스트에 추가
                self.price_sum += self.price[n-1]       # 합계 금액

                print('추가 주문 완료', self.temp, self.menu[n-1], ' : ', self.price[n-1], '원\n', self.price_sum, '원')
            else  :
                if n == 0 :    # 지불을 완료하면
                    print("주문이 완료되었습니다")
                    print(self.order_menu, self.order_price)     # 확인을 위한 리스트 출력해보기
                else :   #  없는 번호 입력할 때
                    print("없는 메뉴입니다")

    # 지불
    def pay(self) :
        # 합계 금액 출력
        print(f'총 합계 금액은 {self.price_sum}원 입니다')
        # 지불 방법
        self.pay_method = 0
        while self.pay_method != 1 and self.pay_method != 2 :    # 1,2 번을 입력하지 않으면 계속 반복
            self.pay_method = int(input('현금 결제는 1번, 카드 결제는 2번을 입력해주세요. : '))
            if self.pay_method == 1:
                print('직원을 호출하겠습니다.')
            elif self.pay_method == 2:
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
            else:
                print('다시 결제를 시도해 주세요')

    # 외곽선
    def table(self):
        outline1 = "⟝" + "-" * 30 +"⟞"
        outline2 = "|" + " " * 31 + "|"

        #외곽 출력
        print(outline1)
        for i in range(5):
            print(outline2)

        # 주문 상품명 : 해당 금액
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i])

        # 합계 금액
        print('합계 금액 :', self.price_sum)

        # 외곽 출력
        for i in range(5):
            print(outline2)
        print(outline1)

a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
